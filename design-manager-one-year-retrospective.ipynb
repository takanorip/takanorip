# デザインマネージャーになって1年を振り返る

こんにちは。takanoripです。

2025年7月1日に株式会社カンムでデザインマネージャーになってから、気づけば1年が経ちました。

就任したときに「デザインマネージャー就任とデザイン組織のこれから」という記事を書き、半年くらい経ったタイミングでも一度振り返りを書きました。今回はさらに半年経って、デザインマネージャーとしての1年をどう捉えているのか、今何を考えているのかを書いてみようと思います。

## デザインエンジニアからマネージャーになって1年

カンムに入社した当初、僕はフロントエンドエンジニアとして入社しました。気持ちとしてはデザインエンジニアとして、デザインと実装の間にある課題を解いていくぞ、という感じでした。

そこから縁あってデザイナーUnitのマネージャーになったわけですが、正直なところ、最初から「マネージャーとしてこういうキャリアを歩みたい」と明確に考えていたわけではありません。

むしろ、自分の強みは手を動かして成果物をつくることにあると思っていたので、マネージャーという仕事に対しては不安のほうが大きかったです。自分が直接つくるのではなく、チームが成果を出せる状態をつくる。その切り替えがちゃんとできるのか、当時はあまり自信がありませんでした。

ただ1年やってみて思うのは、デザインエンジニアとしてやってきたこととデザインマネージャーの仕事は、やっぱり地続きだったということです。

デザインエンジニアは、デザインと実装の間にある摩擦を見つけて、仕組みやルールやツールで解決していく仕事だと思っています。デザインマネージャーも同じように、デザイナーがより良い成果を出すための環境やプロセスを整えていく仕事です。

手を動かして画面をつくる時間は減りましたが、デザインがより良い形でプロダクトや事業に届くための「通り道」をつくる仕事は増えました。自分の中では、この1年でその感覚がかなり強くなった気がします。

## この1年でやってきたこと

就任時の記事では、デザインマネージャーとしてやっていきたいこととして、主に次のようなことを書いていました。

- チームでデザインをする体制づくり
- デザイナーの役割や理想像の定義
- デザインの仕組み化

半年時点では、そこから少し具体化して、デザイナーを作業者から監修者へアップデートすること、フロントエンドエンジニアとデザイナーのワークフローを改善すること、デザイナーの育成に取り組むことを書きました。

1年経った今も、やっていることの大枠はあまり変わっていません。むしろ、この3つは一時的なプロジェクトではなく、デザイン組織をつくっていくうえでずっと向き合い続けるテーマなんだなと感じています。

### デザイナーが全部作業しなくてもよい状態をつくる

この1年で特に意識していたのは、デザイナーがすべての作業を抱え込まなくても、一定の品質で物事が前に進む状態をつくることでした。

少人数のデザイン組織では、どうしても「デザイナーが頑張る」で成立してしまう場面が多くなります。もちろん、それで前に進むこと自体は悪いことではありません。少人数のチームでは、個人の越境や頑張りがプロダクトを前進させる大きな力になります。

ただ、それだけに依存し続けると組織はスケールしません。特定の誰かが見ているから品質が保たれる、特定の誰かが間に入っているからプロジェクトが進む、という状態は短期的には強くても、長期的には不安定です。

なので、デザイナーが毎回手を動かさなくても、一貫性のある判断ができるようにする。デザイナーが全部のボールを拾わなくても、チームとして必要な判断や作業が進むようにする。そういう状態を少しずつ増やしていくことを意識していました。

### ワークフローと役割を言語化する

そのために重要だったのが、ワークフローや役割の言語化です。

以前「越境から協業へ、スケールする組織づくり」という記事でも書きましたが、越境は目的ではなく結果だと思っています。本来は、みんなが自分の専門性を発揮できる状態が一番良いはずです。それでも必要があるから、結果として越境が発生する。

つまり、まずやるべきなのは「もっと越境しよう」と言うことではなく、誰が何に責任を持ち、どのタイミングで誰と相談し、どう意思決定するのかを整理することです。

デザイナーとフロントエンドエンジニアの接続、プロダクト開発におけるデザインの入り方、ハンドオフの粒度、レビューのタイミング。こういったものを少しずつ整理していくことで、個人の頑張りに依存しない協業の形が見えてきました。

役割を定義することは、人を縛ることではありません。むしろ、期待値を明確にして、お互いに安心して協業するための土台だと思っています。

「これは誰の仕事なのか」が曖昧な状態では、誰かが気を利かせ続ける必要があります。でも、それは気を利かせられる人に負荷が偏るということでもあります。役割やプロセスを定義することは、そういう見えにくい負荷を組織の仕組みに移していくことなのだと思います。

### デザインの仕組み化を進める

もうひとつ大きかったのは、デザインの仕組み化です。

デザインシステムの整備や共通パーツの切り出し、ウェブサイト運用基盤の見直しなど、デザインを個別の作業ではなく継続的に運用できる仕組みにしていく取り組みを進めてきました。

最近だと、バンドルカードのウェブサイト基盤をStudioへ移行しました。これは単なるツール変更ではなく、デザイナーやマーケターがエンジニアに依存しすぎず、より素早くウェブサイトを改善していくための体制変更でもありました。

エンジニアが作業しなくても改善できる状態をつくることは、エンジニアの仕事を奪うことではありません。むしろ、エンジニアが本当に向き合うべき技術的な課題に集中できるようにすることでもあります。同じように、デザイナーが全部の作業をやらなくてもよい状態をつくることで、デザイナーはより抽象度の高い課題や、サービス全体の体験設計に向き合いやすくなります。

このあたりは、自分がデザインエンジニア出身だからこそ取り組みやすい領域だったのかもしれません。デザインと実装、デザインと運用、デザインと事業の間にある摩擦を見つけて、そこに仕組みをつくる。そういう仕事がこの1年でかなり増えました。

## マネージャーとして難しかったこと

もちろん、うまくいったことばかりではありません。

半年時点の記事でも書きましたが、マネージャーは大変です。1年やっても普通に大変です。

特に難しいのは、人と人の間にあるものを扱うことです。プロセスやルールを整えればすべて解決するわけではありません。どれだけ正しそうな仕組みをつくっても、それを運用するのは人です。人によって得意不得意も違うし、置かれている状況も違うし、受け取り方も違います。

自分はもともと、問題を構造化して、原因を見つけて、解決策を考えるのが好きです。ただ、マネージャーの仕事では、論理だけでは前に進まないことがたくさんあります。

正しいことを言えばよいわけではない。早く結論を出せばよいわけでもない。相手がどう感じているのか、何に困っているのか、今どれくらい余裕があるのかを見ながら、必要なタイミングで必要なコミュニケーションをする必要があります。

これはまだ全然うまくできている自信がありません。自分が喋りすぎてしまうこともあるし、もっと相手の話を引き出せたほうがよかったなと思うこともあります。助け舟を出すタイミングも難しいです。早すぎると本人の機会を奪ってしまうし、遅すぎると負荷が大きくなってしまいます。

このあたりは、これからもずっと向き合っていくことになりそうです。

## この1年で考え方が変わったこと

1年やってみて、自分の中で一番変わったのは「成果物」への向き合い方かもしれません。

以前は、自分が作ったもの、自分が直接関わったものに対して成果を感じやすかったです。画面を作る、コードを書く、資料を作る。そういう目に見えるアウトプットがあると、自分が何かを前に進めた感覚がありました。

でもマネージャーの仕事は、必ずしも自分の名前がつく成果物として現れるわけではありません。チームの誰かが良い意思決定をできた、プロジェクトがスムーズに進んだ、デザイナーがより重要な仕事に集中できるようになった。そういう状態をつくることが成果になります。

最初はこの感覚に慣れませんでした。自分で手を動かしていないと、何もしていないような気持ちになることもあります。

ただ最近は、デザインマネージャーの仕事は「デザインが良くなる確率を上げること」なのかもしれないと思うようになりました。

毎回自分が判断するのではなく、チームが良い判断をしやすくする。毎回自分が修正するのではなく、良いものが自然に生まれやすい環境をつくる。毎回自分が間に入るのではなく、必要な人同士が適切に協業できる状態をつくる。

そう考えると、マネージャーの仕事は直接的ではないけれど、かなりデザインに近い仕事だなと思います。

## これからやっていきたいこと

この1年で、カンムのデザイン組織は少しずつ形になってきたと思います。ただ、まだまだ始まったばかりです。

これから特にやっていきたいことは、大きく3つあります。

### デザイナーがより上流から関われる状態をつくる

デザイナーが単に画面を作る人ではなく、プロダクトや事業の意思決定により深く関われる状態をつくっていきたいです。

そのためには、デザイナー自身のスキルも必要ですし、組織側の受け入れ態勢も必要です。デザイナーがどのタイミングで何に関わるべきなのか、どのような成果に責任を持つのかを、もっと明確にしていきたいと思っています。

### デザインシステムと運用基盤を育てる

デザインシステムやStudioのような運用基盤は、作って終わりではありません。むしろ、使われ始めてからが本番です。

ルールを増やすことが目的ではなく、プロダクトや事業を前に進めるために必要な判断を減らし、重要な判断に集中できるようにすることが目的です。

デザインシステムもウェブサイト運用も、チームの働き方や事業のフェーズに合わせて育てていく必要があります。ここは引き続き地道にやっていきたいです。

### デザイナーが成長できるチームをつくる

採用ももちろん大事ですが、入社してくれた人が期待以上に活躍できる状態をつくることも同じくらい大事です。

デザイナーの仕事に必要なスキルを整理し、どのような経験を積むと成長できるのか、どのような成果が評価されるのかを明確にしていきたいです。

専門職としてのハードスキルを高めながら、事業や組織にインパクトを出せるデザイナーが増えていく。そういうチームを目指したいと思っています。

## おわりに

デザインマネージャーになって1年、まだまだできていないことも多いですが、少しずつ自分の中でこの仕事の輪郭が見えてきた気がします。

デザインマネージャーは、デザインを直接つくるだけの仕事ではありません。デザインが事業やプロダクトに届くまでの道筋を整え、チームがより良い判断をし、より良い成果を出せる状態をつくる仕事です。

就任したときに書いた「デザインエンジニアのキャリアのロールモデルの1つになりたい」という気持ちは、今も変わっていません。むしろ、デザインエンジニア出身だからこそできるデザインマネジメントがあるのではないか、という感覚はこの1年で強くなりました。

カンムのデザイン組織はまだまだこれからです。いつか「あの組織いいよね」と言ってもらえるような、強くてしなやかなデザイン組織をつくっていけるように、2年目も地道に頑張っていきたいと思います。
